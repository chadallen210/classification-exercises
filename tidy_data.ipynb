from pydataset import data### 1. Attendance Data

Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
attendance_df = pd.read_csv('untidy-data/attendance.csv')
attendance_df.head()
```

```python
attendance_df.shape
```

```python
attendance_df.columns
```

```python
attendance_df.rename(columns={'Unnamed: 0': 'name'}, inplace=True)
attendance_df.head()
```

```python
attendance_melt = attendance_df.melt(id_vars='name', var_name='date', value_name='status')
attendance_melt.head()
```

```python
value = []

for x in attendance_melt.status:
    if x == 'P':
        value.append(1)
    if x == 'H':
        value.append(0.5)
    if x == 'T':
        value.append(0.9)
    if x == 'A':
        value.append(0)

attendance_melt['value'] = value
attendance_melt.head()
```

```python
attendance_percentage = attendance_melt.pivot_table(index='name', aggfunc='mean').reset_index()
attendance_percentage
```

### 3. Cake Recipes

#### a. Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

```python
cake_recipes_df = pd.read_csv('untidy-data/cake_recipes.csv')
cake_recipes_df
```

#### b. Tidy the data as necessary.

```python
cake_recipes_df[['recipe', 'position']]= cake_recipes_df['recipe:position'].str.split(':', expand=True)
cake_recipes_df
```

```python
cake_recipes_df = cake_recipes_df.drop(columns='recipe:position')
cake_recipes_df
```

```python
cake_recipes_df = cake_recipes_df[['recipe', 'position', '225', '250', '275', '300']]
cake_recipes_df
```

```python
cake_recipes_tidy = cake_recipes_df.melt(id_vars=['recipe', 'position'], var_name='temperature', value_name='value')
cake_recipes_tidy.head()
```

#### c. Which recipe, on average, is the best? recipe b

```python
cake_recipes_tidy.pivot_table(values='value', index='recipe').sort_values(by='value', ascending=False).reset_index()
```

#### d. Which oven temperature, on average, produces the best results? 275

```python
cake_recipes_tidy.pivot_table(values='value', index='temperature').sort_values(by='value', ascending=False).reset_index()
```

#### e. Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
cake_recipes_tidy.nlargest(1, 'value')
```
